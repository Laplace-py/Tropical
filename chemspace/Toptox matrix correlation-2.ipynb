```python
import pandas as pd
df=pd.read_csv('../Downloads/data_vero_human_cells.csv')
df.head()
```

```python
#df.drop(['B', 'C'], axis=1)
df2=df.drop(['InChIKey',"Molecule"], axis=1)
df2.head()
```

```python
corr = df2.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style('darkgrid')
sns.set(font_scale=0.9, rc={'figure.figsize':(12,10)})
sns.heatmap(corr, cmap='RdYlBu_r', vmin=-1, vmax=1, annot=True)
plt.title('Correlation Heatmap', fontsize=12)
plt.savefig('3d7_corr.png', bbox_inches='tight',
            transparent=True, format='png', dpi=300)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

corr = np.corrcoef(np.random.randn(10, 200))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    
sns.set_style('darkgrid')
sns.set(font_scale=0.9, rc={'figure.figsize':(12,10)})
sns.heatmap(corr, cmap='RdYlBu_r', vmin=-1, vmax=1, annot=True)
plt.title('Correlation Heatmap', fontsize=12)
plt.savefig('3d7_corr.png', bbox_inches='tight',
            transparent=True, format='png', dpi=300)
  
```
